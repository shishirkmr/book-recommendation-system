```python
# Import Libraries
import pandas as pd
import numpy as np
```

```python
# Read Long format DF
long_format_df=pd.read_csv('./data/csv/long_format_df.csv')
```

```python
long_format_df.head()
```

```python
from scipy.sparse import csr_matrix
long_format_df = long_format_df.drop_duplicates(['users', 'book'])
```

```python
book_user_pivot = long_format_df.pivot(index = 'book', columns = 'users', values = 'rating').fillna(0)
```

```python
book_user_pivot.head()
```

```python
book_user_rating_matrix = csr_matrix(book_user_pivot.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(book_user_rating_matrix)
```

```python
# query_index = np.random.choice(long_format_pivot.shape[0])
query_index = 37

print(query_index)
```

```python
distances, indices = model_knn.kneighbors(book_user_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python

book_user_pivot.index[query_index]
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(book_user_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, book_user_pivot.index[indices.flatten()[i]], distances.flatten()[i]))

```

# Find Similar Users

```python
user_book_pivot = long_format_df.pivot(index = 'users', columns = 'book', values = 'rating').fillna(0)
```

```python
user_book_pivot.head()
```

```python
user_book_rating_matrix = csr_matrix(user_book_pivot.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(user_book_rating_matrix)
```

```python
user_book_pivot[user_book_pivot.index=='Tiffany']
```

```python
user_book_pivot[user_book_pivot.index=='Hamza ALi']
```

```python
def find_similar_users(user_name):
    similar_users=[]
    query_index = get_index(user_name)
    if query_index == -1:
        print('{} User Not Found!!'.format(user_name))
    distances, indices = model_knn.kneighbors(user_book_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Similar user for {0}:\n'.format(user_book_pivot.index[query_index]))
        else:
            print('{0}: {1}, with distance of {2}:'.format(i, user_book_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
            if distances.flatten()[i]>0 and distances.flatten()[i]<0.2:
                similar_users.append(user_book_pivot.index[indices.flatten()[i]])
    return similar_users

def get_index(user_name):
    for i in range(0,len(user_book_pivot.index)):
            if user_book_pivot.index[i] == user_name:
                return i
    return -1
```

```python
find_similar_users('Tiffany')
```

```python
similar_users_df=user_book_pivot.index.map(lambda x: {x:find_similar_users(x)})
```
